### **2da Pregunta:** ¿Cuál fue la facturación por categoria em 2023?

**Observación:** Para la facturación, consideraremos aquí el producto entre el precio unitario y la cantidad, sumados al costo de envío de cada producto.


> #### <strong><font color="7b71ff" size=6>Prompt:</font></strong>

Tenemos un DataFrame llamado "df" y queremos construir un gráfico de barras horizontales que muestre la facturación por categoría de Zoop.

En "df", tenemos las siguientes columnas:
* categoria: categorías de los productos
* precio_unitario: precio de una unidad del producto
* cantidad: cantidad vendida del producto
* envio: costo de envío total de la compra.

Deseamos crear una visualización que muestre la facturación de cada categoría, donde la facturación se calcula multiplicando el precio unitario por la cantidad y sumando el costo de envío.

Utiliza las bibliotecas de visualización de datos de Python para crear esta visualización.


````python
# Paso 1: Calcular la facturación por fila
df['facturacion'] = (df['precio_unitario'] * df['cantidad']) + df['envio']
````

````python
# Paso 2: Agrupar por categoría y sumar la facturación total por categoría
facturacion_por_categoria = df.groupby('categoria')['facturacion'].sum().sort_values(ascending=False).reset_index()
````

````python
# Paso 3: Visualización con gráfico de barras horizontales
plt.figure(figsize=(10, 6))
sns.barplot(data=facturacion_por_categoria, x='facturacion', y='categoria', palette='viridis',orient= 'h')

plt.title('Facturación por Categoría de Producto', fontsize = 16)
plt.xlabel('Facturación Total',fontsize = 12)
plt.ylabel('Categoría',fontsize = 12)
plt.tight_layout()
plt.show()
````

<font color="7b71ff" size=4> Comentarios: </font>

---

Aquí especificamos el tipo de gráfico, guiando a ChatGPT en la creación de nuestro visual. Podemos notar que en un gráfico de barras el enfoque está más centrado en las categorías, además de ser una forma más fluida de leer textos que podrían ser demasiado extensos para un gráfico de columnas.

###**4a Pregunta:** ¿Cómo está dividida la venta por trimestre por método de pago? ¿Existe algún período en el que algún tipo de pago se destaque?


> #### <strong><font color="7b71ff" size=6>Prompt:</font></strong>

Tenemos un DataFrame llamado "df" y queremos construir un gráfico de barras apiladas que muestre las ventas por trimestre en relación con el método de pago de Zoop.

En "df", tenemos las siguientes columnas:
* fecha: con las fechas de las compras en formato datetime (`aaaa-mm-dd`)
* facturacion: con los valores de las ventas
* metodo_pago: con el método de pago elegido en la compra

Deseamos crear una visualización que muestre las ventas por trimestre, agrupando los métodos de pago de cada trimestre correspondientes a cada compra.

Utiliza las bibliotecas de visualización de datos de Python para crear esta visualización.


````python
# 1. Crear una nueva columna 'trimestre' que extrae el trimestre de la columna 'fecha'
df['trimestre'] = df['fecha'].dt.to_period('Q')

# 2. Agrupar los datos por trimestre y método de pago, sumando la facturación
ventas_trimestre = df.groupby(['trimestre', 'metodo_pago'])['facturacion'].sum().unstack().fillna(0)

# 3. Crear el gráfico de barras apiladas
ventas_trimestre.plot(kind='bar', stacked=False, figsize=(10, 6), colormap='viridis')

# 4. Añadir etiquetas y título
plt.title('Ventas por Trimestre y Método de Pago', fontsize=16)
plt.xlabel('Trimestre', fontsize=12)
plt.ylabel('Facturación Total', fontsize=12)

# Rotar las etiquetas del eje x para mayor claridad
plt.xticks(rotation=0)

# 5. Mostrar el gráfico
plt.legend(title='Método de Pago por trimestre')
plt.tight_layout()
plt.show()
````

<font color="7b71ff" size=4> Comentarios: </font>

---

El gráfico de columnas apiladas añade a la comparación entre períodos o categorías, la comparación también entre los datos de cada período/categoría. Es como si pusiéramos una "lupa" en cada columna y visualizáramos la composición de cada una de acuerdo con la variable deseada (aquí, las formas de pago).

Utiliza las bibliotecas de visualización de datos de Python para crear esta visualización.


````python
def ventas_trimestre():

  # 1. Crear columna de trimestre
  df['trimestre'] = df['fecha'].dt.to_period('Q')

  # 2. Agrupar por trimestre y método de pago
  ventas_trimestre = df.groupby(['trimestre', 'metodo_pago'])['facturacion'].sum().unstack().fillna(0)

  # 3. Crear gráfico
  fig, ax = plt.subplots(figsize=(10, 6))
  fig.patch.set_facecolor(GRIS_1)
  ax.set_facecolor(GRIS_1)

  # 4. Dibujar barras apiladas
  ventas_trimestre.plot(
      kind='bar',
      stacked=True,
      ax=ax,
      color=[BLANCO, GRIS_3, GRIS_1, GRIS_3],
      width=0.7,
      edgecolor='black'
  )

  # 5. Título
  ax.set_title('Ventas por Trimestre y Método de Pago en 2023', fontsize=18, color=GRIS_5)

  # 6. Estilo etiquetas
  ax.set_xlabel(None)
  ax.set_ylabel(None)
  ax.tick_params(axis='x', labelsize=12, colors=AZUL_1, rotation=0)
  ax.tick_params(axis='y', labelsize=12, colors=AZUL_1)

  # 7. Formato de eje Y como "MXN$ X M" sin decimales
  def formato_mxn_millon(valor, _):
      return f'MXN$ {valor/1e6:.0f} M'

  ax.yaxis.set_major_formatter(ticker.FuncFormatter(formato_mxn_millon))

  # 8. Eliminar bordes
  for spine in ['top', 'bottom', 'left', 'right']:
      ax.spines[spine].set_visible(False)

  # 9. Cuadrícula punteada solo en Y
  ax.grid(axis='y', linestyle=':', color=GRIS_3, linewidth=1)

  # 10. Leyenda
  ax.legend(title='Método de Pago por trimestre de 2023', frameon=False)

  plt.tight_layout()
  return plt
````

````python
def proporcion_cashback():

  # Valores de participación (por ejemplo: resultado de df['cashback'].value_counts())
  participacion_cashback = df['cashback'].value_counts()

  # Colores por orden del índice ("No", "Sí")
  colores = [GRIS_3 if label == 'No' else AZUL_1 for label in participacion_cashback.index]

  # Crear figura y gráfico
  fig, ax = plt.subplots(figsize=(6, 6))
  fig.patch.set_facecolor(GRIS_1)
  ax.set_facecolor(GRIS_1)

  # Donut chart
  wedges, texts, autotexts = ax.pie(
      participacion_cashback,
      labels=participacion_cashback.index,
      autopct='%1.1f%%',
      startangle=180,
      colors=colores,
      pctdistance=0.8,
      wedgeprops={'width': 0.4}
  )

  # Estilo de etiquetas y porcentajes
  for text, color in zip(texts, colores):
      text.set_color(color)
      text.set_fontsize(12)

  for autotext in autotexts:
      autotext.set_color(BLANCO)
      autotext.set_fontsize(12)

  # Título
  ax.set_title('¿Participan en el programa de cashback?', fontsize=18, color=GRIS_5)

  # Círculo proporcional
  ax.axis('equal')

  plt.tight_layout()
  return plt
````
